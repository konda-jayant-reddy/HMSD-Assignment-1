```python
import geopandas as gpd
import rasterio
import matplotlib.pyplot as plt
import numpy as np
from rasterio.features import geometry_mask
from rasterio.mask import mask
from rasterio.windows import Window
```

```python
# plot the cdne43a_v3r1/cdne43a.tif DEM file
with rasterio.open('cdne43a_v3r1/cdne43a.tif') as src:
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.imshow(src.read(1), cmap='pink')
    ax.set_title('DEM')
    plt.show()
```

```python
# plot the BMC_Wards.shp shapefile
wards = gpd.read_file('BMC_Wards.shp')
fig, ax = plt.subplots(figsize=(10, 10))
wards.plot(ax=ax)
ax.set_title('Wards')
plt.show()
```

```python
# clip the DEM data using the shapefile
with rasterio.open('cdne43a_v3r1/cdne43a.tif') as src:
    out_image, out_transform = mask(src, wards.geometry, crop=True)
    out_meta = src.meta.copy()
    out_meta.update({"driver": "GTiff",
                     "height": out_image.shape[1],
                     "width": out_image.shape[2],
                     "transform": out_transform})
    with rasterio.open("clipped.tif", "w", **out_meta) as dest:
        dest.write(out_image)

# plot the clipped DEM data
with rasterio.open('clipped.tif') as src:
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.imshow(src.read(1), cmap='pink')
    ax.set_title('Clipped DEM')
    plt.show()
```

```python
# create a drainage direction map by applying the D8 algorithm
with rasterio.open('clipped.tif') as src:
    dem = src.read(1)
    rows, cols = dem.shape
    d8 = np.zeros((rows, cols), dtype=np.int8)
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            d8[i, j] = 0
            if dem[i, j] > dem[i - 1, j]:
                d8[i, j] += 1
            if dem[i, j] > dem[i - 1, j + 1]:
                d8[i, j] += 2
            if dem[i, j] > dem[i, j + 1]:
                d8[i, j] += 4
            if dem[i, j] > dem[i + 1, j + 1]:
                d8[i, j] += 8
            if dem[i, j] > dem[i + 1, j]:
                d8[i, j] += 16
            if dem[i, j] > dem[i + 1, j - 1]:
                d8[i, j] += 32
            if dem[i, j] > dem[i, j - 1]:
                d8[i, j] += 64
            if dem[i, j] > dem[i - 1, j - 1]:
                d8[i, j] += 128

# plot the drainage direction map
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(d8, cmap='jet')
ax.set_title('Drainage Direction Map')
plt.show()
```

```python
# Open the DEM file
with rasterio.open('clipped.tif') as src:
    # Read metadata and full DEM
    transform = src.transform
    full_dem = src.read(1)
    
    # Define the coordinates (x, y) of the center of the sub-region
    x_center, y_center = 400, 400  # replace these with your coordinates
    
    # Define the radius (in number of cells)
    radius = 100
    
    # Calculate the window to read based on radius
    col_off = x_center - radius
    row_off = y_center - radius
    width = height = 2 * radius
    
    # Read the sub-region
    window = Window(col_off=col_off, row_off=row_off, width=width, height=height)
    sub_dem = src.read(1, window=window)
    sub_dem
    # Plot the sub-region
    plt.imshow(sub_dem, cmap='jet')
    plt.title('Sub-region from DEM')
    plt.show()

# save the sub-region as a GeoTIFF file
with rasterio.open('sub_dem.tif', 'w', **src.meta) as dst:
    dst.write(sub_dem, 1)
```

```python
# create a flow accumulation map for the sub-region
with rasterio.open('sub_dem.tif') as src:
    dem = src.read(1)
    rows, cols = dem.shape
    d8 = np.zeros((rows, cols), dtype=np.int8)
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            d8[i, j] = 0
            if dem[i, j] > dem[i - 1, j]:
                d8[i, j] += 1
            if dem[i, j] > dem[i - 1, j + 1]:
                d8[i, j] += 2
            if dem[i, j] > dem[i, j + 1]:
                d8[i, j] += 4
            if dem[i, j] > dem[i + 1, j + 1]:
                d8[i, j] += 8
            if dem[i, j] > dem[i + 1, j]:
                d8[i, j] += 16
            if dem[i, j] > dem[i + 1, j - 1]:
                d8[i, j] += 32
            if dem[i, j] > dem[i, j - 1]:
                d8[i, j] += 64
            if dem[i, j] > dem[i - 1, j - 1]:
                d8[i, j] += 128

# plot the flow accumulation map
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(d8, cmap='jet')
ax.set_title('Flow Accumulation Map')
plt.show()
```

```python
# plot a flood inundation map for the sub-region 

with rasterio.open('sub_dem.tif') as src:
    # Read metadata and full DEM
    transform = src.transform
    full_dem = src.read(1)

    # define the flood parameters
    flood_level = 100
    flood_depth = 10

    # create a flood inundation map
    flood_map = np.where(full_dem > flood_level, flood_depth, 0)

    # plot the flood inundation map
    plt.imshow(flood_map, cmap='pink')
    plt.title('Flood Inundation Map')
    plt.show()
```

```python
# calculate peak discharge for the sub-region using Q = C*I*A
# where C is the runoff coefficient, I is the rainfall intensity, and A is the area of the sub-region

with rasterio.open('sub_dem.tif') as src:
    # for each cell
    dem = src.read(1)
    runoff = 0.2 * (1 - np.exp(-0.2 * dem))
    cell_area = src.res[0] * src.res[1]
    rainfall_intensity = 0.9  # in m/s
    # for each cell in the sub-region use the formula Q = C*I*A
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            dem[i, j] = runoff[i, j] * rainfall_intensity * cell_area

    # plot the peak discharge map
    plt.imshow(dem, cmap='pink')
    plt.title('Peak Discharge Map')
    plt.show()
```

```python
# calculate runoff coeffecient for the sub-region
with rasterio.open('sub_dem.tif') as src:
    # use the formula: runoff = 0.2 * (1 - exp(-0.2 * slope))
    # calculate the slo
    slope = src.read(1)
    # calculate the runoff
    runoff = 0.2 * (1 - np.exp(-0.2 * slope))
    # plot the runoff map
    plt.imshow(runoff, cmap='pink')
    plt.title('Runoff Coefficient Map')
    plt.show()
```
